```python
import pandas as pd
import numpy as np

df = {'Fix_ID' : [1, 2, 3, 4], 'Team_1' : ['A', 'C', 'B', 'A'], 'Team_2' : ['B', 'D', 'A', 'B'], '1_G' : [5, 2, 0, 1], '2_G' : [5, 1, 3, 1], 'ALFIE' : [6.0, 3.0, 2.0, 4.0]}
fixture_df = pd.DataFrame(df)
fixture_df
```

```python
swapped_df = fixture_df.rename(columns={'Team_1' : 'Team_2', 'Team_2' : 'Team_1', '1_G' : '2_G', '2_G' : '1_G'})
swapped_df
```

```python
stacked_df = pd.concat([fixture_df, swapped_df], ignore_index=True).rename(columns={"Team_1" : "Team_10", "Team_2" : "Team_20"})
stacked_df
```

```python
grouped_df = stacked_df.groupby(['Team_1', 'Team_2']).agg({
    '1_G' : "sum",
    '2_G' : "sum",
    'ALFIE' : "mean",
    'Team_1' : "count"
})
grouped_df
```

```python
grouped_df['Total_goals'] = grouped_df['1_G'] + grouped_df['2_G']
grouped_df
```
