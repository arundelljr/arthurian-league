```python
import pandas as pd
import numpy as np

df = pd.read_csv("C:/Users/arund/coding/arthurian-league/data/02_intermediate/ALFIE_fixtures.csv")
df
```

```python
stats_df = df[['Home_Team', 'Away_Team', 'Home_goals', 'Away_goals', 'ALFIE_rating']]
stats_df
```

```python
grouped_df = stats_df.groupby(['Home_Team', 'Away_Team']).agg({
    "Home_goals": "sum",
    "Away_goals": "sum",
    "ALFIE_rating": "mean",
    "Home_Team" : "count"
}).rename(columns={"Home_Team" : "Total_games"}).reset_index()
grouped_df
```

```python
merged_df = grouped_df.merge(grouped_df, left_on=['Home_Team', 'Away_Team'], right_on=['Away_Team', 'Home_Team'], how='left', suffixes=('_1', '_2'))
filtered = merged_df[merged_df['Home_Team_1'] == 'Arthurian League Representative Team']
filtered
```

```python
fill_df = merged_df.fillna(0)

# Create a cleaned DataFrame with the desired columns
cleaned_df = fill_df.assign(
    Team_1=fill_df['Home_Team_1'],
    Team_2=fill_df['Away_Team_1'],
    Team_1_goals=fill_df['Home_goals_1'] + fill_df['Away_goals_2'],
    Team_2_goals=fill_df['Away_goals_1'] + fill_df['Home_goals_2'],
    ALFIE_rating=(
        fill_df['ALFIE_rating_1'] * fill_df['Total_games_1'] +
        fill_df['ALFIE_rating_2'] * fill_df['Total_games_2']
    ) / (fill_df['Total_games_1'] + fill_df['Total_games_2']),
    Total_games=fill_df['Total_games_1'] + fill_df['Total_games_2']
)[['Team_1', 'Team_2', 'Team_1_goals', 'Team_2_goals', 'ALFIE_rating', 'Total_games']]

cleaned_df['Team_1_goals'] = cleaned_df['Team_1_goals'].astype('Int64') 
cleaned_df['Team_2_goals'] = cleaned_df['Team_2_goals'].astype('Int64')
cleaned_df['Total_games'] = cleaned_df['Total_games'].astype('Int64')

cleaned_df
```
