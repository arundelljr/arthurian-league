# Add in rudementary 'Fixture Rating' based on closeness of a game, also taking into account the total goals scored 
# Arthurian League Fixture Index Evaluation (ALFIE)
import numpy as np

def _ALFIE_rating(homescore, awayscore):
    ttl_gls = homescore + awayscore
    gl_diff = np.abs(homescore - awayscore)
    ttl_gls_rt = np.sqrt(ttl_gls/2)
    
    close_rating = np.where(gl_diff < 3.5, 3.5 - gl_diff, 
                                np.where(gl_diff > 4.5, -2, 0)) 

    ALFIE = np.abs(round(ttl_gls_rt + close_rating))
    return ALFIE

```python
import pandas as pd

al_results_df = pd.read_csv(r"C:\Users\arund\Prospect_week\al_results_normal.csv")
al_results_df.dtypes
```

```python
al_results_df['ALFIE_rating'] = _ALFIE_rating(al_results_df['Homegoals'], al_results_df['Awaygoals'])
al_results_df.head()
```
